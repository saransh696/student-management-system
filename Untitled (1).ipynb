```python
x=[ ]
while True:
    y=input("enter task")
    if y =="add student":
        a=input("enter name")
        b=int(input("enter roll number"))
        c=int(input("enter marks"))
        def add_student():
            z=f"{a}\n{b}\n{c}"
            x.append(z)
        add_student()
        print("student added successfully")
    elif y =="view student":
        def view_student():
            if len(x)==0:
                print("no student found")
            else:
                for i in x:
                    print(i)
        view_student()
    elif y=="delete student":
        def delete_student():
            m=int(input("enter student number"))
            del x[m]
        delete_student()
        print("deleted successfully")
    elif y=="exit":
        break
    else:
        print("error")
        
```
